```python
def get_messy_index(df_size, n_indices, seed=None):
    """Return random indices to inject messy data."""
    if seed is not None:
        np.random.seed(seed)
    return np.random.randint(0, df_size, n_indices)


def create_table(df, cols, key, table_name=None, path=None, messy_data=None, seed=None):
    """
    Create a clean or intentionally messy CSV table.

    Parameters:
    - df: source DataFrame
    - cols: list of columns to include
    - key: column(s) to drop duplicates on
    - table_name: output CSV file name (no extension)
    - path: output folder path
    - messy_data: list of dicts describing how to mess up the data
    - seed: optional random seed for reproducibility
    """
    table = df[cols].drop_duplicates(subset=key).copy()
    table.reset_index(drop=True, inplace=True)
    
    if messy_data:
        for messy in messy_data:
            col = messy.get('col')  # optional, not needed for 'duplicates'
            n_indices = min(messy.get('messy_rows_count', 0), table.shape[0])
            indices = get_messy_index(table.shape[0], n_indices, seed)

            if messy['type'] == 'nan':
                table.loc[indices, col] = np.nan

            elif messy['type'] == 'inconsistent':
                typos = messy.get('typos', {})
                table.loc[indices, col] = table.loc[indices, col].apply(
                    lambda x: typos.get(str(x), x) if pd.notnull(x) else x
                )

            elif messy['type'] == 'lowercase':
                table.loc[indices, col] = table.loc[indices, col].apply(
                    lambda x: x.lower() if pd.notnull(x) else x
                )
                
            elif messy['type'] == 'uppercase':
                table.loc[indices, col] = table.loc[indices, col].apply(
                    lambda x: x.upper() if pd.notnull(x) else x
                )

            elif messy['type'] == 'duplicates':
                # Number of duplicates to add per selected row
                dup_factor = messy.get('dup_factor', 1)

                # Select the rows to duplicate
                rows_to_duplicate = table.iloc[indices]

                # Repeat them
                duplicated_rows = pd.concat([rows_to_duplicate] * dup_factor, ignore_index=True)

                # Optional shuffle for realism
                if messy.get('shuffle', True):
                    duplicated_rows = duplicated_rows.sample(frac=1, random_state=seed).reset_index(drop=True)

                # Append duplicates to the table
                table = pd.concat([table, duplicated_rows], ignore_index=True)

                print(f"Added {len(duplicated_rows)} duplicate rows")

            else:
                print(f"Warning: Unknown messy type '{messy['type']}' — skipped.")


    # Save final version
    table.to_csv(f'{path}/{table_name}.csv', index=False, encoding='utf-8-sig')
    print(f"Saved table '{table_name}' with {len(table)} rows to {path}")
    return table
```

# Customers

```python
messy_customers = [
    {
        'type': 'inconsistent',
        'col': 'Customer Segment',
        'messy_rows_count': 500,
        'typos': {'Consumer': ' consumer ', 'Home Office': ' home office '}
    },
    {
        'type': 'lowercase',
        'col': 'Customer Fname',
        'messy_rows_count': 10000
    },
    {
        'type': 'lowercase',
        'col': 'Customer Lname',
        'messy_rows_count': 10000
    },
    {
        'type': 'duplicates',
        'messy_rows_count': 112,  # Select 100 random customers
        'dup_factor': 1           # Add 1 duplicate for each
    }
]

customers_df = create_table(
    df=df,
    cols = ['Customer Id', 'Customer Fname', 'Customer Lname', 'Customer Email', 'Customer Password', 'Customer Segment', 'Customer Zipcode'],
    key = ['Customer Id'],
    table_name = 'customers',
    path = 'CRM',
    messy_data=messy_customers,
    seed=42
)
```
